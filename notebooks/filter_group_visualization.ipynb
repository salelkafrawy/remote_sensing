```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from pathlib import Path
```

```python
imgs_path = "/home/mila/s/sara.ebrahim-elkafrawy/scratch/ecosystem_project/remote_sensing/notebooks/images/"
```

```python
# visualize the first conv layer filters
plt.figure(figsize=(20, 17))
for i in np.arange(32):
    plt.subplot(6, 6, i+1) #  because in conv0 we have 3x3 filters and total of 32 (see printed shapes)
    img = Image.open(Path(imgs_path) / f"input_for_mosaik_layer0_featmap{i}.jpg")
    plt.imshow(img)
    plt.axis('off')
    plt.rcParams['savefig.facecolor']='white'
    plt.savefig('./images/mosaik_conv1_most_activated_featmaps.png')
plt.show()
```

```python
# visualize the 2nd conv layer filters
plt.figure(figsize=(20, 17))
for i in np.arange(64):
    plt.subplot(8, 8, i+1) # (8, 8) because in conv1 we have 3x3 filters and total of 64 (see printed shapes)
    img = Image.open(Path(imgs_path) / f"input_for_mosaik_layer3_featmap{i}.jpg")
    plt.imshow(img)
    plt.axis('off')
    plt.rcParams['savefig.facecolor']='white'
    plt.savefig('./images/mosaik_conv2_most_activated_featmaps.png')
plt.show()
```
